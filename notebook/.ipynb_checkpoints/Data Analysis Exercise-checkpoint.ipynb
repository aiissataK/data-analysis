```python
import pandas as pd
```

## Loading datasets

### Bids.csv
- bid_id - unique id for this bid
- bidder_id – Unique identifier of a bidder (same as the bidder_id used in train.csv and test.csv)
- auction – Unique identifier of an auction
- merchandise –  The category of the auction site campaign, which means the bidder might come to this site by way of searching for "home goods" but ended up bidding for "sporting goods" - and that leads to this field being "home goods". This categorical field could be a search term, or online advertisement. 
- device – Phone model of a visitor
- time - Time that the bid is made (transformed to protect privacy).
- country - The country that the IP belongs to
- ip – IP address of a bidder (obfuscated to protect privacy).
- url - url where the bidder was referred from (obfuscated to protect privacy). 

```python
bids = pd.read_csv("../data/bids.csv")
print("Data dimension")
print("Number of observations: "+str(bids.shape[0]))
print("Number of features: "+str(bids.shape[1]))
bids.head()
```

### Train.csv
- bidder_id – Unique identifier of a bidder.
- payment_account – Payment account associated with a bidder. These are obfuscated to protect privacy. 
- address – Mailing address of a bidder. These are obfuscated to protect privacy. 
- outcome – Label of a bidder indicating whether or not it is a robot. Value 1.0 indicates a robot, where value 0.0 indicates human. 
- The outcome was half hand labeled, half stats-based. There are two types of "bots" with different levels of proof:
    1. Bidders who are identified as bots/fraudulent with clear proof. Their accounts were banned by the auction site.
    2. Bidder who may have just started their business/clicks or their stats exceed from system wide average. There are no clear proof that they are bots. 

```python
train = pd.read_csv("../data/train.csv")
print("Data Dimension")
print("Number of observations: "+str(train.shape[0]))
print("Number of features: "+str(train.shape[1]))
train.head()
```

### Merging datasets

We first join the two datasets to build a unique one that will contain all the information of bidders from those two datasets

```python
df = pd.merge(bids, train,how='inner', on='bidder_id')
df.head()
```

## Missing values

Then, we handle missing values.
We notice that all features have the same number of observations but one, which is the country.

```python
df.count()
```

Let us see how many are missing.

```python
df.isnull().sum()
```

And in percentage.

```python
100*df.isnull().sum()/df.shape[0]
```

There are not a lot of missing observations for the feature country.<br/>
Instead of deleting them, we will impute the value by a new one so that all null values be replaced by the term 'not specified'. <br/>
We keep the same feature and create another one that will contain the imputed values.


```python
df['country_imputed'] = df['country'].fillna('not specified')
df.isnull().sum()
```

Now we see that the new feature country_imputed has no missing value

 ## Data exploration and data analysis 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Distribution of bids by bidder type

```python
df['outcome'].value_counts().plot(kind='bar')
plt.title("Barplot : Outcome")
plt.xlabel("Human and Robots")
plt.ylabel("Number")
plt.show()
```

The figure above shows the number of bids done by human bidder and robot bidder. We can see that most of the bids are done by human but the number of bids executed by robots is significant.

## Distribution of bids by day

The time feature is not very illustrative. So we will extract new interesting features like date and hour that might help us in our analysis.

```python
df['date'] = pd.to_datetime(df['time']).dt.date
df['hour'] = pd.to_datetime(df['time']).dt.hour
df.head()
```

Now that we have the dates and hours, we will see the number of bids sent every hour in each day.

```python
f, ax = plt.subplots(figsize=(13, 7))
sns.countplot(x="hour", hue="date",data=df);
plt.show()
```

This figure above shows the number of bids sent in each hour every day. We can notice that in each day there are bids in one time interval and the interval may change each day.
Now let's see the number of bids day by day.

```python
for date in df['date'].unique():
    df['hour'][df['date'] == date].value_counts().plot(kind='bar')
    plt.title("Barplot :"+ str(date))
    plt.xlabel("Hours")
    plt.ylabel("Count")
    plt.show()
```

- In each figure above we can see the interval of time when bids were sent. Therefore, we can assume that there were active auctions during these hours.

## Distribution of bids by country

There is a lot of country in the dataset to plot them all. Our solution is to group them by continent. 
For that, we create a module that returns the continent of each country and its full name.

```python
module = __import__('countries.countryinfo')
```

We create a new feature 'continent' which is the continent for the corresponding country

```python
l=[]
for country in df['country_imputed']:
    if (country=='not specified'):
        l.append('not specified')
    else :
        l.append(module.countryinfo.getContinent(country))
se = pd.Series(l)
df['continent'] = se.values
```

 We create a new feature 'country_name' which is the full name for the corresponding country

```python
l=[]
for country in df['country_imputed']:
    if (country=='not specified'):
        l.append('not specified')
    else :
        l.append(module.countryinfo.getCountry(country))
se = pd.Series(l)
df['country_name'] = se.values
```

```python
del(l)
del(se)
```

Now let's see the distribution of bids by continent and by human and robots

```python
f, ax = plt.subplots(figsize=(13, 7))
sns.countplot(x="continent", hue="outcome",data=df);
plt.title("Barplot : Continent")
plt.xlabel("Nom des Continents")
plt.ylabel("Count")
plt.show()
```

We notice that bids from human and robots come more from Asia with more than 1.2 millions from humans and 200.000 from robots which is one million less. 
However, in Oceania, even if there are not a lot of bids from there, the amount of bids from humans and from robots is the same. In North America, the number of bids from robots is half the number from humans. In Europe, it's almost the same.


Let's see from which country there are more robot bidders than human bidders

```python
a = pd.crosstab(df['country_name'], df['outcome'])
a[a[1] > a[0]]
```

In percentage

```python
a = pd.crosstab(df['country_name'], df['outcome']).apply(lambda x: 100 *(x / x.sum()), axis=1)
a[a[1] >50]
```

These two tables show the list of countries that have more bids from robot. Germany has the most robot bidder:18530.
However with percentage, Japan take the leads with 94% of bids coming from robot then South Korea with 90%.

## Which ip addresses are commonly used by robot for each country? 

```python
a = df[df['outcome']==1]
a = a[['bidder_id','ip','country_name','continent']].groupby(['continent','country_name','ip'])['bidder_id'].agg(['count'])
a['somme'] = a.groupby(['country_name']).transform('sum')
a['pourcentage'] = 100 * a['count'] / a['somme']
a[(a['count']>2) &(a['pourcentage']>50)]
```

In this table, we show countries where the same ip address is used more than 50% times by robot. There are more african countries. Turkmenistan takes the lead with an ip address used almost 98% by robots.
